```python
import geopandas as gpd
```

```python
gdf = gpd.read_file(r"D:\MINNIE\contest_gyeongnam_data_analysis\data\LARD_ADM_SECT_SGG_경남\LARD_ADM_SECT_SGG_48_202505.shp")
```

```python
gdf
```

```python
gdf_merge = gdf.dissolve(by='COL_ADM_SE')
```

```python
gdf_merge
```

```python
gdf_merge.loc[gdf_merge["SGG_NM"] == "경상남도 창원시 의창구", "SGG_NM"] = "경상남도 창원시"
gdf_merge
```

```python
gdf_merge["SGG_NM_cleaned"] = gdf_merge["SGG_NM"].str.replace("경상남도", "", regex=False).str.strip()
gdf_merge
```

```python
gdf_merge.crs
```

```python
gdf.plot()
```

```python
gdf_merge.plot()
```

```python
gdf_merge = gdf_merge.to_crs("EPSG:4326")
```

```python
gdf_merge.to_file("GN_map.geojson", driver="GeoJSON")
```

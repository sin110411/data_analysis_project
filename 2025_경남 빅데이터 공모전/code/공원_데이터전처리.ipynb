문화행사

```python
df_festival = pd.read_csv(r"D:\MINNIE\contest_gyeongnam_data_analysis\data\문화행사.csv", encoding='euc-kr')
df_festival.head(5)
```

```python
df_festival = df_festival[df_festival['시군명'].isin(['산청군', '창원시', '김해시', '의령군'])]
df_festival.head(5)
```

```python
df_festival.info()
```

```python
df_festival['행사시작일'] = pd.to_datetime(df_festival['행사시작일'], errors='coerce')
df_festival['행사종료일'] = pd.to_datetime(df_festival['행사종료일'], errors='coerce')

df_festival.info()
```

```python
df_festival['행사일수'] = (df_festival['행사종료일'] - df_festival['행사시작일']).dt.days + 1
df_festival.head()
```

```python
df_festival = df_festival.groupby("시군명")["행사일수"].sum().reset_index()
df_festival = df_festival.rename(columns={"행사일수": "총행사일수"})
```

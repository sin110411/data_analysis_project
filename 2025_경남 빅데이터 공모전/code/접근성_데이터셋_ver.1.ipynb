```python
#전국 주차장

pk = pd.read_csv('전국주차장정보표준데이터.csv', encoding = 'cp949')
```

```python
pk
```

```python
print(pk.columns)

pk = pk[['제공기관명','주차장명','위도', '경도',]]

pk
```

```python
pk_li = pk['제공기관명'].to_list()
```

```python
시군구명 = []

for i in pk_li:
    if len(i.split())>=2:
        시군구명.append(i.split()[1])
    else:
        시군구명.append(i.split()[0])
```

```python
pk['시군구명'] = 시군구명
# pk = pk.drop('제공기관명',axis = 1)
```

```python
# print(pk['시군구명'].value_counts())

pk = pk[pk['시군구명'].isin(경남)]

pk
```

```python
print(pk['시군구명'].value_counts())
```
